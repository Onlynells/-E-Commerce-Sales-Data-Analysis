## Importing Libararies

```python
# Importing libraries
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plt
```

## Loading the Dataset

```python
#load the dataset
df= pd.read_csv('/content/amazon.csv')
```

We have only 2 missing value found in the rating_count column.

```python
#Dropping the rows containig the missing values
df=df.dropna()
```

```python
#confirming if there is any missing value
df.isnull().sum()
```

## Data Cleaning

```python
#previewing the discounted price and actual price
df[['discounted_price', 'actual_price']].head()
```

```python
# sampling the unique values
df['discounted_price'].unique()[:5]
```

```python
# converting the columns to numeric format
# Removing '₹' and ',' then converting to float
df['discounted_price'] = df['discounted_price'].str.replace('₹','').str.replace(',','').astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹','').str.replace(',','').astype(float)
```

```python
#Viewing the discounted price and actual price colums after conversion
df[['discounted_price', 'actual_price']].head()
```

```python
#Convert rating_count to integer values (remove commas if present).
df['rating_count'] = df['rating_count'].str.replace(',','').astype(int)
```

```python
#cheking the data types
df.dtypes
```

Discounted_price and actual_price converted to float.
rating_count converted to integer.

```python
# Extracting the unique product categories
unique_categories = df['category'].unique()
unique_categories
```

```python
#no of counts per category
category_counts = df['category'].value_counts()
category_counts.head()
```

```python
# Extracting the main category (first part before '>')
df['main_category'] = df['category'].str.split('|').str[0].str.strip()

# Extracting the sub-category (second part if available)
df['sub_category'] = df['category'].str.split('|').str[1].str.strip()
```

```python
# Count the main categories
main_category_counts = df['main_category'].value_counts()
main_category_counts
```

```python
# Count the sub categories
sub_category_counts = df['sub_category'].value_counts()
sub_category_counts
```

```python
# Creating a detailed category structure with main and sub categories
df['product_category'] = df['main_category'] + ' | ' + df['sub_category']
product_category_counts = df['product_category'].value_counts()
product_category_counts
```

## Data Transformation

```python
# Creating a discount_amount column
df['discount_amount'] = df['actual_price'] - df['discounted_price']
df['discount_amount'] .head()
```

```python
#viewing the discount_percentage column
df['discount_percentage'].head()
```

```python
#Convert discount_percentage to numeric by removing '%' and converting to float
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)
df['discount_percentage'].head()
```

```python
df.columns
```

No date/timestamp column is present in the dataset

## Explanatory Data Analysis

```python
#distribution of ratings using a histogram.
plt.figure(figsize=(8,5))
plt.hist(df['rating'], bins=10, edgecolor='black', alpha=0.7)
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.title("Distribution of Ratings")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Correlation between discount_percentage and rating
# Convert columns to numeric (handle errors)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
correlation = df[['discount_percentage', 'rating']].corr()
correlation
```

The correlation value is -0.1557, which is a weak negative correlation.
This suggests that higher discounts tend to be associated with slightly lower ratings, but the relationship is not strong

```python
# Group by main_category and sum the rating_count to get an estimate for sale
category_sales = df.groupby('product_category')['rating_count'].sum().sort_values(ascending=False)
category_sales.head(10)

```

```python
# Plot the top 10 categories
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.barplot(x=category_sales.head(10).index, y=category_sales.head(10).values)
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Product Categories by Total Rating Count', fontsize=16)
plt.xlabel('Product Category', fontsize=14)
plt.ylabel('Rating Count', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Assuming 'category_sales' is already sorted in descending order
top_5_categories = category_sales.head(5)

# Create pie chart
plt.figure(figsize=(10, 6))
plt.pie(top_5_categories.values, labels=top_5_categories.index,
        autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

# Add title
plt.title("Top 5 Product Categories by  Rating Count", fontsize=14)

# Show plot
plt.show()
```

```python
# Analyze the relationship between discount and rating_count
# Create a scatter plot between discount_amount and rating_count_numeric
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='discounted_price', y='rating_count', alpha=0.6)
plt.title('Relationship between Discount and Rating Count', fontsize=16)
plt.xlabel('Discount Amount', fontsize=14)
plt.ylabel('Rating Count', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation between discount_amount and rating_count_numeric
corr = df[['discounted_price', 'rating_count']].corr()
corr
```

The value -0.027261 is very close to 0, indicating a weak or no correlation between discounted_price and rating_count.
The negative sign suggests that as discounted_price increases, rating_count slightly decreases, but the effect is very weak.
